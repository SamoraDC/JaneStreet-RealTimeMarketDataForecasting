# Jane Street Conservative RLS Late Submission

Attach the official competition data and the private Dataset containing `jane-street-conservative-rls-late-submission`. Keep internet disabled and use GPU if available. Submit the saved Kaggle Notebook version, not a local parquet or zip file.

```python
from pathlib import Path
import runpy


def find_package_dir() -> Path:
    input_root = Path('/kaggle/input')
    matches = []
    for candidate in sorted(input_root.glob('*')):
        if (
            (candidate / 'submission/submission.py').exists()
            and (candidate / 'artifacts/jane_street_submission/base_models/manifest.json').exists()
            and (candidate / 'artifacts/jane_street_submission/meta_rls_experts_alpha10000_f0p995/rls_state.npz').exists()
        ):
            matches.append(candidate)
    if not matches:
        raise FileNotFoundError(
            'Attached Dataset not found. Attach the extracted private Dataset '
            'jane-street-conservative-rls-late-submission, not just a zip file.'
        )
    if len(matches) > 1:
        print('Multiple candidate packages found; using:', matches[0])
    return matches[0]


package_dir = find_package_dir()
print('Using package:', package_dir)
runpy.run_path(str(package_dir / 'kaggle_notebook_launcher.py'), run_name='__main__')
```
